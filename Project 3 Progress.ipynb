## Checking for Duplicate Rows

```python
duplicate_rows_df = Maindata[Maindata.duplicated()] 
duplicate_rows_df.shape
```

```python
def Detecting_Outliers(Data):
    
    threshold=1
    mean = np.mean(Data) # computing mean of the dataset
    std =np.std(Data)  # computing the standard deviation
    
    for y in Data:
        z_score= (y - mean)/std 
        if np.abs(z_score) > threshold:
            Outliers.append(y)        
    return Outliers

fig=plt.figure()

for i  in range(0,len(Maindata.columns)):
    Outliers=[]
   
    if Maindata[Maindata.columns[i]].dtypes== np.int64 or Maindata[Maindata.columns[i]].dtypes== np.int64  :
        
        Outliers=Detecting_Outliers(Maindata.iloc[:,i])
        if len(Outliers)>0:
                   
            ax = plt.subplot(1,1,1)
            #Variable
            ax.boxplot(Maindata[list(Maindata.columns)[i]])
            plt.title(list(Maindata.columns)[i] + " Outlier Plot")
            plt.show()
```
